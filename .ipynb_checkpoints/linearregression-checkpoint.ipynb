```python
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
```

```python
features,target=make_regression(1000,n_features=4,noise=0.2,
                                coef=False,n_informative=2,n_targets=1,random_state=42)



```

```python
features.shape
```

```python
target.shape
```

```python
target
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
regression=LinearRegression()
```

```python
regression
```

```python
model=regression.fit(features,target)
```

```python
regression
```

```python
model
```

```python
model==regression
```

```python
model.intercept_
```

```python
model.coef_
```

## model.coef_ will give the parameters of the linear regression line
## model.intercept_ will give the intercept of the linear regression line

```python
target[0]
```

```python
model.predict(features)[0]
```

```python
## to print the score of the model (i.e, how well the model get performed)
```

```python
model.score(features,target)
```
